```python
import os
from PIL import Image
import textwrap

dataset_path = 'drive/MyDrive/PlantVillage'

class_names = sorted(os.listdir(dataset_path))


num_classes = len(class_names)
plt.figure(figsize=(min(20, num_classes * 2), 4))

for i, class_name in enumerate(class_names):
    class_folder = os.path.join(dataset_path, class_name)

    if not os.path.isdir(class_folder):
        continue

    image_files = [f for f in os.listdir(class_folder) if f.lower().endswith(('png', 'jpg', 'jpeg'))]
    if not image_files:
        continue

    image_path = os.path.join(class_folder, image_files[0])
    image = Image.open(image_path)

    plt.subplot(1, num_classes, i + 1)
    plt.imshow(image)
    plt.axis('off')

    title = "\n".join(textwrap.wrap(class_name, 12))
    plt.title(title, fontsize=8)

plt.tight_layout()
plt.show()

```

# train baseline model

```python
from fastai.vision.all import *

def train_and_export_baseline(base_path,
                     model=resnet18,
                     img_size=128,
                     epochs=1,
                     bs=64,
                     export_n='model.pkl'):
    if torch.cuda.is_available():
        device = torch.device('cuda')
        print("Using GPU :D")
    else:
        device = torch.device('cpu')
        print("Using CPU :(")

    base_path = Path(base_path)

    dblock = DataBlock(
        blocks=(ImageBlock, CategoryBlock),
        get_items=get_image_files,
        splitter=GrandparentSplitter(train_name='train', valid_name='val'),
        get_y=parent_label,
        item_tfms=Resize(img_size)

    )
    dls = dblock.dataloaders(base_path,bs=bs , num_workers=8, seed=42)
    dls.to(device)


    learn = vision_learner(dls, model, metrics=[Precision(average='macro'), Recall(average='macro'), F1Score(average='macro')])
    learn.to_fp16()
    learn.model.to(device)
    learn.fine_tune(epochs)

    test_files = get_image_files(base_path/'test')
    test_dl = dls.test_dl(test_files, with_labels=True)
    _, precision, recall, f1 = learn.validate(dl=test_dl)
    print(f"Precision, Recall, F1: {precision:.4f}, {recall:.4f}, {f1:.4f}")

    learn.export(export_n)
    print(f"Model exported :]")

    return learn


```

```python
from fastai.vision.all import *
from pathlib import Path

def train_and_export_imp_baseline(base_path,
                     model=resnet18,
                     img_size=256,
                     epochs=10,
                     lr=2e-3,
                     bs=64,
                     export_n='model.pkl'):
    if torch.cuda.is_available():
        device = torch.device('cuda')
        print("Using GPU :D")
    else:
        device = torch.device('cpu')
        print("Using CPU :(")

    base_path = Path(base_path)

    dblock = DataBlock(
        blocks=(ImageBlock, CategoryBlock),
        get_items=get_image_files,
        splitter=GrandparentSplitter(train_name='train', valid_name='val'),
        get_y=parent_label,
        item_tfms=Resize(img_size),
        batch_tfms=[*aug_transforms(
            flip_vert=True,
            max_rotate=10,
            max_zoom=1.1,
            max_lighting=0.2,
            max_warp=0.2,
            p_affine=0.75,
            p_lighting=0.75),
        Normalize.from_stats(*imagenet_stats)]
    )
    dls = dblock.dataloaders(base_path,bs=bs , num_workers=8, seed=42)
    dls.to(device)


    learn = vision_learner(dls, model, metrics=[Precision(average='macro'), Recall(average='macro'), F1Score(average='macro')])
    learn.to_fp16()
    learn.model.to(device)
    learn.fine_tune(epochs, base_lr=lr)

    test_files = get_image_files(base_path/'test')
    test_dl = dls.test_dl(test_files, with_labels=True)
    _, precision, recall, f1 = learn.validate(dl=test_dl)
    print(f"Precision, Recall, F1: {precision:.4f}, {recall:.4f}, {f1:.4f}")

    learn.export(export_n)
    print(f"Model exported :]")

    return learn


```

```python
imp_baseline_model_resnet18 = train_and_export_imp_baseline('C:\\Users\\User\\Downloads\\Plant disease classification\\plant model data_split',
                           model=resnet18,
                           img_size=256,
                           epochs=10,
                           lr=2e-3,
                           bs=64,
                           export_n='imp_baseline_resnet18.pkl')
```
